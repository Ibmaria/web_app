```python
#New york latitude , longitude range
boundary={'min_lng':-74.5,
              'min_lat':40.7,
              'max_lng':-72.8, 
              'max_lat':41.8}
```

```python
##  consider locations within New York City
for long_value in ['pickup_longitude', 'dropoff_longitude']:
  df_taxi = df_taxi[(df_taxi[long_value] > boundary['min_lng']) & (df_taxi[long_value] < boundary['max_lng'])]
for lat_value in ['pickup_latitude', 'dropoff_latitude']:
  df_taxi = df_taxi[(df_taxi[lat_value] > boundary['min_lat']) & (df_taxi[lat_value] <boundary['max_lat'])]
```

https://stackoverflow.com/questions/27928/calculate-distance-between-two-latitude-longitude-points-haversine-formula#:~:text=from%20math%20import%20cos%2C%20asin,*R*asin...

```python
def distance_trip(latitude1,longitude1,latitude2,longitude2):
  r = 6373 # earth's radius
  latitude1 = np.deg2rad(latitude1)
  longitude1 = np.deg2rad(longitude1)
  latitude2 = np.deg2rad(latitude2)
  longitude2= np.deg2rad(longitude2)
  dlat = latitude2 - latitude1
  dlon = longitude2 - longitude1
  a = np.sin(dlat/2)**2 + np.cos(latitude1) * np.cos(latitude2) * np.sin(dlon/2)**2
  c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
  distance = r*c
  return distance
def direction_angle(latitude1,longitude1,latitude2,longitude2):
  dlon = longitude2 - longitude1
  x = np.cos(latitude2)* np.sin(dlon)
  y= np.cos(latitude1)* np.sin(latitude2) - np.sin(latitude1)*np.cos(latitude2) * np.cos(dlon)
  beta_en_radians = np.arctan2(x,y)
  beta_en_degres = np.rad2deg(beta_en_radians)
  return beta_en_degres

```

```python
#distance away from  three major airports, Central park and Manhattan

airports = {'JFK': (-73.78,40.643),'LGA': (-73.87, 40.77),'EWR' : (-74.18, 40.69),'MNT':(-73.97,40.7831),'Cenpark':(-73.96,40.77)}
for airport in airports:
  df_taxi['pickup_dist_' + airport] = distance_trip(df_taxi['pickup_latitude'], df_taxi['pickup_longitude'], airports[airport][1], airports[airport][0])
  df_taxi['dropoff_dist_' + airport] = distance_trip(df_taxi['dropoff_latitude'], df_taxi['dropoff_longitude'], airports[airport][1], airports[airport][0]) 
        
```

```python
#trip distance
df_taxi['trip_distance'] = distance_trip(df_taxi['pickup_latitude'], df_taxi['pickup_longitude'], df_taxi['dropoff_latitude'], df_taxi['dropoff_longitude'])
```

```python
#trip direction north south east west
df_taxi['direction'] = direction_angle(df_taxi['pickup_latitude'], df_taxi['pickup_longitude'], df_taxi['dropoff_latitude'], df_taxi['dropoff_longitude'])
```

```python
#Convert pickup_datetime from Object to Datetime object
df_taxi['pickup_datetime']=pd.to_datetime(df_taxi['pickup_datetime'],format='%Y-%m-%d %H:%M:%S UTC')
```

```python
weeks=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
months=['January','February','March','April','May','June','July','August','September','October','November','December']
hours_names=['hour1','hour2','hour3','hour4','hour5','hour6','hour7','hour8','hour9','hour10','hour11','hour12','hour13','hour14','hour15','hour16','hour17','hour18','hour19','hour20','hour21','hour22','hour23','hour24']
```

```python
df_taxi['day_of_week'] = df_taxi['pickup_datetime'].dt.dayofweek
df_taxi['month'] = df_taxi['pickup_datetime'].dt.month
df_taxi['hour'] = df_taxi['pickup_datetime'].dt.hour
```

```python
peak_time = df_taxi.groupby(['hour']).size()

print('Fare by hour of the day:')
print(peak_time)
```

```python
def peak_hour(row):
  if ((row.hour >= 6 and row.hour <= 9) or (row.hour >= 16 and row.hour <= 20)):
    return 1
  else:
    return 0
```

```python
df_taxi['peak_hour'] = df_taxi.apply(lambda row: peak_hour(row), axis = 1)
```

```python
def is_weekend(row):
  if (row.day_of_week == 5 or row.day_of_week == 6):
    return 1
  else:
    return 0
def is_weekday(row):
  if (row.day_of_week != 5 or row.day_of_week != 6):
    return 1
  else:
    return 0

```

```python
df_taxi['is_weekend'] = df_taxi.apply(lambda row: is_weekend(row), axis = 1)
```

```python
df_taxi['is_weekday'] = df_taxi.apply(lambda row: is_weekday(row), axis = 1)
```

```python
df_taxi['day_of_week'].replace(to_replace=[i for i in range(0,7)],value=weeks,inplace=True)
df_taxi['month'].replace(to_replace=[i for i in range(1,13)],value=months,inplace=True)
df_taxi['hour'].replace(to_replace=[i for i in range(0,24)],value=hours_names,inplace=True)
```

```python
df_taxi
```

```python
week_dummies=pd.get_dummies(df_taxi['day_of_week'])
months_dummies =pd.get_dummies(df_taxi['month'])
hours_dummies =pd.get_dummies(df_taxi['hour'])
df_taxi=pd.concat([df_taxi,week_dummies,months_dummies,hours_dummies],axis=1)
```

```python
df_taxi.head()
```

```python
### function that convert latitudes and longtitudes in degres to radians format
df_taxi['pickup_latitude'] = np.deg2rad(df_taxi['pickup_latitude'].values)
df_taxi['pickup_longitude'] = np.deg2rad(df_taxi['pickup_longitude'].values)
df_taxi['dropoff_latitude'] = np.deg2rad(df_taxi['dropoff_latitude'].values)
df_taxi['dropoff_longitude'] = np.deg2rad(df_taxi['dropoff_longitude'].values)
df_taxi['direction'] = np.deg2rad(df_taxi['direction'].values)
```
